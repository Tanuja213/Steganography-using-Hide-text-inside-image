```python
import numpy as np
```

```python
d={chr(i):i for i in range(255)}  #characters to ASCII
c={i:chr(i) for i in range(255)}  #Ascii to character
```

```python
#message and encrypted key
text="Hello"
key="key"
```

```python
#random pixel value Image
x =np.random.randint(0,255,(10,10,3),dtype=np.uint8)
```

```python
x_enc=x.copy()
n=0
m=0
z=0
l=len(text)
kl=0
for i in range(l):
    char_val=d[text[i]]^d[key[kl]]
    for bit_pos in range(8):
        bit=(char_val >> (7-bit_pos)) & 1
        org_val=x_enc[n,m,z]
        # Use bitwise AND with 254 (binary 11111110) to clear the last bit,
        # which is equivalent to org_val & ~1 but avoids negative numbers.
        x_enc[n,m,z]=(org_val & 254)| bit
       # print(f"Embedding bit {bit} of '{text[i]}'  at ({n},{m},{z})  original={orig_val} new={x_enc[n,m,z]}")
        z=(z+1)%3
        if z == 0:
            m=m+1
            if m == x_enc.shape[1]:
                m=0
                n=n+1
    kl=(kl+1)%len(key)
```

```python
n,m,z=0,0,0
kl=0
l=len(text)
decrypt=""
for i in range(l):
    val=0
    for bit_pos in range(8):
        bit= x_enc[n,m,z] & 1
        val=(val << 1) | bit
        print(f"Reading bit {bit} from ({n},{m},{z}")
        z=(z+1)%3
        if z==0:
            m=m+1
            if m==x_enc.shape[1]:
                m=0
                n=n+1
    orig_char= c[val ^ d[key[kl]]]
    decrypt+=orig_char
    print(f"Decrypted byte: {val} XOR {d[key[kl]]} = {val ^d[key[kl]]} -> '{orig_char}'")
    kl=(kl+1)%len(key)
print("Decrypted text :", decrypt)
```
